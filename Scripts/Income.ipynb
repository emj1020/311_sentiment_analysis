```python
gdf['County Name'] = gdf['County Name'].replace({'Staten Island': ' Richmond', 'Manhattan': ' New York', 'Brooklyn': ' Kings',  'Queens': ' Queens',  'Bronx': ' Bronx'})
```

```python
gdf
```

```python
m = pd.merge(gdf, inc, on='County Name')
```

```python
m
```

```python
m.shape
```

```python
m.columns
```

```python
m = m.drop(['NTAAbbrev', 'NTAType', 'CDTA2020', 'CDTAName', 'Tract', 'State'], axis=1)
```

```python
m
```

```python
m =  m.groupby(['NTAName'], as_index=False).mean()
```

```python
m
```

```python
grouped = m.groupby('NTAName')

weighted_avg = grouped.agg({'Median Income': lambda x: np.average(x, weights=m.loc[x.index, 'Population'])})

weighted_avg = weighted_avg.rename(columns={'Median Income': 'Weighted Avg Income'})

result = m.merge(weighted_avg, on='NTAName')
```

```python
result.to_csv('Income.csv', index=False)
```
