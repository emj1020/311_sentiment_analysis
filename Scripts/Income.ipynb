```python
m = m.drop(['BoroCode_left', 'NTA2020_left',
       'NTAAbbrev', 'CDTA2020_left',
       'Shape_Leng_left', 'Shape_Area_left', 'geometry', 'index_right', 'BoroCode_right', 'BoroName_right', 'CT2020', 'BoroCT2020',
       'CDEligibil', 'NTA2020_right', 'CDTA2020_right',
       'CDTANAME'], axis=1)
```

```python
m
```

```python
merged =  m.groupby(['NTAName_left'], as_index=False).mean()
```

```python
merged
```

```python
# Define the custom aggregation function to calculate the weighted average
def weighted_average(group):
    weights = group['Population']
    values = group['Median Income']
    return (weights * values).sum() / weights.sum()

# Group the data by NTAName_left and apply the custom aggregation function
weighted_avg_df = m.groupby('NTAName_left').apply(weighted_average)

# Print the resulting DataFrame
weighted_avg_df
```

```python
wa = pd.DataFrame(weighted_avg_df)
wa
```

```python
wa.columns
```

```python
wa = wa.rename(columns={0: "Weighted Median Income"})
```

```python
wa = wa.reset_index()
```

```python
income = pd.merge(merged, wa, on='NTAName_left')
```
